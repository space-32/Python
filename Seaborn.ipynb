```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips= sns.load_dataset('tips')
#tips is in-built dataset is available in seaborn
tips
```

```python
tips.head()
```

```python
tips.tail()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
#Joint point
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#pairplot
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#rug plot
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips)
```

Categorical Plots


```python
#bar plot and count plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
#box plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='rainbow')
```

```python
#violinplot
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='rainbow')
```

```python
#stripplot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#swarmplot
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker',palette='rainbow')
```
