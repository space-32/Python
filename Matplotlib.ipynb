```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([0,6])
ypoints=np.array([0,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,8])
ypoints=np.array([2,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,8])
ypoints=np.array([2,10])
plt.plot(xpoints,ypoints,'o') #plotting without a line
plt.show()
```

```python
#Multiple points
x=np.array([1,2,6,9])
y=np.array([2,7,1,10])
plt.plot(x,y)
```

```python
#x=np.array([1,2,6,9])
#if we dont specify the x then default values will be 0,1,2,3
y=np.array([2,7,1,10])
plt.plot(y)
```

Matplotlib Markers

```python
y=np.array([3,8,2,9])
plt.plot(y,marker='o',mec='r',mfc='g',ms=10)
#mfc=marker face color, mec=marker edge color, ms=marker size
plt.show()
```

Matplotlib Lines

```python
y=np.array([3,8,2,9])
plt.plot(y,linestyle='dotted',color='y')
#Instead of dotted we can write :, for das --
plt.show()
```

```python
y=np.array([3,8,2,9])
plt.plot(y,linestyle='solid',color='orange',lw=10)
#lw= line width
plt.show()
```

```python
y1=np.array([3,6,8,10])
y2=np.array([6,2,7,12])
plt.plot(y1)
plt.plot(y2)
```

Matplotlib Labels


```python
x=np.array([80,85,86,90,92,95,89])
y=np.array([100,110,120,130,135,140,145])
plt.plot(x,y)
plt.xlabel('Marks in 1st Exam')
plt.ylabel('Marks in 2nd Exam')
plt.show()
```

```python
x=np.array([80,85,86,90,92,95,89])
y=np.array([100,110,120,130,135,140,145])
plt.plot(x,y)
plt.xlabel('Marks in 1st Exam')
plt.ylabel('Marks in 2nd Exam')
plt.grid()
plt.show()
```

Subplots in Matplotlib


```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
#1- no. od rows, 2- no. of cols, 1-plot no. respectively
plt.plot(x,y)

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.plot(x,y)

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()
```

Matplotlib Plots


```python
#Scatter plot
x=np.array([80,85,86,90,92,95,89])
y=np.array([100,110,120,130,135,140,145])
plt.scatter(x,y)
plt.show()
```

```python
#Bar plot
plt.bar(x,y)
```

```python
plt.barh(x,y, color='c')#for horizontal
```

```python
#Histograph
x=np.random.normal(170,10,250)
plt.hist(x)
```

```python
#Pie Chart
x=np.array([35,25,25,15])
plt.pie(x)
```
